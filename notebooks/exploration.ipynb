```python
# Checking data completeness
df = pd.read_csv("../data/goteborg-femman-air-quality.csv", parse_dates=['date'], skipinitialspace=True)
df_aq = df[["date", "pm25"]]
```

```python
plt.figure(figsize=(10,6))

plt.plot(df['date'], df['pm25'])
plt.ylabel('PM2.5')
plt.xlabel('Date')

plt.grid(alpha=0.3)

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['pm25'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of PM2.5 Levels')
plt.xlabel('PM2.5')
plt.ylabel('Frequency')
plt.grid(alpha=0.3)
plt.show()
```

```python
# Boxplot to see outliers (dots), max and min values (whiskers), interquartile range (box) and median
import seaborn as sns

plt.figure(figsize=(10,6))
sns.boxplot(x=df['date'].dt.month, y=df['pm25'], palette='coolwarm')
plt.title('Monthly PM2.5 Distribution')
plt.xlabel('Month')
plt.ylabel('PM2.5 (g/m)')
plt.show()
```
